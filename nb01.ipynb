```python
# Bibliotecas
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from IPython.display import Image
```

```python
# 1 - Leitura de imagem com OpenCV
imagem = cv2.imread("/home/imagens/datacience.jpg")
cv2_imshow(imagem)
```

```python
# 2 - Array matricial da imagem
imagem = cv2.imread("/home/imagens/datacience.jpg")
print(imagem)
```

```python
# 3 - Apresentação de imagem com  Matplotlib
plt.imshow(imagem)
```

```python
# 4 - Leitura da imagem com escala de cinza
imagem_cinza = cv2.imread("/home/imagens/datacience.jpg",0)

# Array matricial
print(imagem_cinza)

# Visualização da imagem.
plt.imshow(imagem_cinza,cmap='gray')
```

```python
# 5 - Imagem negativa
img_soccer = cv2.imread("/home/imagens/soccer.jpg")
img_soccer_reversed = (255-img_soccer)
plt.imshow(img_soccer_reversed)
```

```python
# 6 Decomposição RGB
img_bgr = cv2.imread("/home/imagens/python.png",cv2.IMREAD_COLOR)
b,g,r = cv2.split(img_bgr)

# Exibição dos Canais
plt.figure(figsize=[20,5])
plt.subplot(141);plt.imshow(r,cmap='gray');plt.title("Red Channel");
plt.subplot(142);plt.imshow(g,cmap='gray');plt.title("Green Channel");
plt.subplot(143);plt.imshow(b,cmap='gray');plt.title("Blue Channel");

# Mescla dos Canais
imgMerged = cv2.merge((b,g,r))

# Saida da mescla
plt.subplot(144);plt.imshow(imgMerged[:,:,::-1]);plt.title("Merged Output");
```

```python
# BGR vs RGB
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
# 7 Decomposição HSV
img_hsv = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2HSV)

# Divisão B,G,R
h,s,v = cv2.split(img_hsv)

# Exibição dos Canais
plt.figure(figsize=[20,5])
plt.subplot(141);plt.imshow(h,cmap='gray');plt.title("H Channel");
plt.subplot(142);plt.imshow(s,cmap='gray');plt.title("S Channel");
plt.subplot(143);plt.imshow(v,cmap='gray');plt.title("V Channel");
plt.subplot(144);plt.imshow(img_hsv);plt.title("Original");
```

```python
h_new = h+10
img_HSV_merged = cv2.merge((h_new,s,v))
img_rgb = cv2.cvtColor(img_HSV_merged, cv2.COLOR_HSV2RGB)

# Show the channels
plt.figure(figsize=[20,5])
plt.subplot(141);plt.imshow(h,cmap='gray');plt.title("H Channel");
plt.subplot(142);plt.imshow(s,cmap='gray');plt.title("S Channel");
plt.subplot(143);plt.imshow(v,cmap='gray');plt.title("V Channel");
plt.subplot(144);plt.imshow(img_rgb);plt.title("Modified");
```
