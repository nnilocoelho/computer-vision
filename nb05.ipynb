```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from IPython.display import Image
import argparse
```

```python
# 1 - Suavização por média
# Justificatica, média utilizada para a imagem não sofrer muito desfoque.
image = cv2.imread('/home/imagens/fogo.jpeg', cv2.IMREAD_UNCHANGED)
cv2_imshow(image)
kernelSizes = [(2, 2), (4, 4), (8, 8)]

# loop over the kernel sizes
for (kX, kY) in kernelSizes:
	# apply an "average" blur to the image using the current kernel
	# size
	blurred = cv2.blur(image, (kX, kY))
	cv2_imshow(blurred)
	
	cv2.waitKey(0)
	cv2.destroyAllWindows()
```

```python
# 2 - Gaussiana
# Justificativa: remoção de ruidos na imagem.
image = cv2.imread('/home/imagens/agua.jpeg', cv2.IMREAD_UNCHANGED)
cv2_imshow(image)

# loop over the kernel sizes again
for (kX, kY) in kernelSizes:
	# apply a "Gaussian" blur to the image
    blurred = cv2.GaussianBlur(image, (kX, kY), 0)
    cv2_imshow(blurred)
    
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
# 3 - Suavização Média
# Justificativa: remoção de todos os ruidos da imagem, forma mais eficaz
# diferente da Gaussiana.

image = cv2.imread('/home/imagens/floresta.jpeg', cv2.IMREAD_UNCHANGED)
cv2_imshow(image)

# loop over the kernel sizes a final time
for k in (3, 9, 15):
	# apply a "median" blur to the image
	blurred = cv2.medianBlur(image, k)
	cv2_imshow(blurred)
	
	cv2.waitKey(0)
	cv2.destroyAllWindows()
```

```python
# Suavização Bilateral
# Justificativa: preservação das bordas
image = cv2.imread('/home/imagens/ar.jpeg', cv2.IMREAD_UNCHANGED)
cv2_imshow(image)
params = [(11, 21, 7), (11, 41, 21), (11, 61, 39)]

# loop over the diameter, sigma color, and sigma space
for (diameter, sigmaColor, sigmaSpace) in params:
	# apply bilateral filtering to the image using the current set of
	# parameters
	blurred = cv2.bilateralFilter(image, diameter, sigmaColor, sigmaSpace)
  
  #output
	cv2_imshow(blurred)
	cv2.waitKey(0)
	cv2.destroyAllWindows()
```
